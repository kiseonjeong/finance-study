```python
import finterstellar as fs
```

```python
df = fs.get_price('NVDA', start_date='2023-01-01', end_date='2023-03-31')
print(df)
```

```python
fs.draw_chart(df, right='NVDA')
```

```python
fs.rsi(df, w=14)
fs.draw_chart(df, left='rsi', right='NVDA')
```

```python
fs.indicator_to_signal(df, factor='rsi', buy=40, sell=60)
```

```python
fs.position(df)
fs.draw_chart(df, left='rsi', right='position_chart')
```

```python
fs.evaluate(df, cost=.001)
fs.draw_chart(df, left='acc_rtn_dp', right='NVDA')
```
